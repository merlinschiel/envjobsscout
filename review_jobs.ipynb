```python
import pandas as pd
import os
import ipywidgets as widgets
from IPython.display import display, IFrame, HTML, clear_output

CURATED_DB = 'curated_portfolio.csv'

# Create the curated database if it doesn't exist
if not os.path.exists(CURATED_DB):
    db_columns = ['Company', 'Job_Type_or_Title', 'Location', 'Source', 'Link', 'Notes', 'Status']
    pd.DataFrame(columns=db_columns).to_csv(CURATED_DB, index=False)
    print("✅ Created new curated database: curated_portfolio.csv")
else:
    print("✅ Curated database loaded.")
```

```python
if os.path.exists('job_map.html'):
    display(HTML("<h3>🗺️ Visual Job Scout</h3><p><i>Explore the map, then use the tool below to add jobs to your database.</i></p>"))
    display(IFrame(src='./job_map.html', width='100%', height=500))
else:
    print("❌ Map not found. Did the script in Cell 2 run successfully?")
```

```python
# Load the freshly scraped jobs
if os.path.exists('geoeco_jobs_clean.csv'):
    df_scraped = pd.read_csv('geoeco_jobs_clean.csv')
    
    # Create a list of options for the dropdown (Company - Title)
    job_options = [f"{row['Company']} | {row['Title']}" for index, row in df_scraped.iterrows()]
    
    print("📥 EXTRACT JOBS FROM SCRAPER")
    job_selector = widgets.Dropdown(options=job_options, description='Select Job:', layout={'width': 'max-content'})
    add_scraped_btn = widgets.Button(description="Add to Curated DB", button_style='success')
    scraped_output = widgets.Output()

    def save_selected_job(b):
        with scraped_output:
            scraped_output.clear_output()
            # Find the selected job in the scraped dataframe
            selected_str = job_selector.value
            comp, title = selected_str.split(" | ", 1)
            
            # Extract row data
            job_data = df_scraped[(df_scraped['Company'] == comp) & (df_scraped['Title'] == title)].iloc[0]
            
            # Append to curated DB
            new_entry = pd.DataFrame([{
                'Company': job_data['Company'], 'Job_Type_or_Title': job_data['Title'], 
                'Location': job_data['Location'], 'Source': 'Greenjobs Script', 
                'Link': job_data['Link'], 'Notes': f"Skills: {job_data['Skills']}", 'Status': 'To Review'
            }])
            new_entry.to_csv(CURATED_DB, mode='a', header=False, index=False)
            print(f"✅ Successfully added '{title}' at {comp} to your database!")

    add_scraped_btn.on_click(save_selected_job)
    display(widgets.HBox([job_selector, add_scraped_btn]))
    display(scraped_output)
else:
    print("❌ Scraped data not found.")
```

```python
print("✍️ ADD MANUAL ENTRY (General Company or Outside Job)")

man_company = widgets.Text(description="Company:")
man_title = widgets.Text(description="Job/Field:")
man_location = widgets.Text(description="Location:")
man_link = widgets.Text(description="Link:")
man_notes = widgets.Text(description="Notes:")
add_man_btn = widgets.Button(description="Save Manual Entry", button_style='info')
man_output = widgets.Output()

def save_manual_entry(b):
    with man_output:
        man_output.clear_output()
        new_entry = pd.DataFrame([{
            'Company': man_company.value, 'Job_Type_or_Title': man_title.value, 
            'Location': man_location.value, 'Source': 'Manual/Networking', 
            'Link': man_link.value, 'Notes': man_notes.value, 'Status': 'Watchlist'
        }])
        new_entry.to_csv(CURATED_DB, mode='a', header=False, index=False)
        print(f"✅ Added {man_company.value} to your database!")
        # Clear fields
        man_company.value = ''; man_title.value = ''; man_location.value = ''; man_link.value = ''; man_notes.value = ''

add_man_btn.on_click(save_manual_entry)
display(widgets.VBox([man_company, man_title, man_location, man_link, man_notes, add_man_btn, man_output]))
```
